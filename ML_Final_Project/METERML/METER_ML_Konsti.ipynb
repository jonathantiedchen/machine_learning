```python
import geopandas as gpd

# Read GeoJSON file
gdf = gpd.read_file('/Users/konstantinreicheneder/Desktop/METER-ML/test_dataset.geojson')

# Extract Image_Folder and Type attributes
table_data = gdf[['Image_Folder', 'Type']]
```

```python
table_data
```

```python
table_data['Image_Folder'] = table_data['Image_Folder'].apply(lambda x: 'samples/' + x + '.png')
```

```python
table_data
```

```python
types = ['Negative', 'ProcessingPlants', 'WWTreatment', 'Landfills', 'RefineriesAndTerminals', 'CAFOs', 'Mines']
filtered_data = table_data[table_data['Type'].isin(types)]
filtered_data
```

```python
from sklearn.preprocessing import LabelEncoder

# Create an instance of LabelEncoder
encoder = LabelEncoder()

# Fit and transform the 'Type' column
filtered_data['Type_encoded'] = encoder.fit_transform(filtered_data['Type'])

# Display the updated table_data
filtered_data
```

```python
type_counts = filtered_data['Type_encoded'].value_counts()
print(type_counts)
```

```python
filtered_data.to_csv('/Users/konstantinreicheneder/Desktop/METER-ML/METER_ML_test.csv', index=False)
```
