```python
import geopandas as gpd
import pandas as pd
```

```python
mapping = {
    "CAFOs": [1, 0, 0, 0, 0, 0, 0],
    "Landfills": [0, 1, 0, 0, 0, 0, 0],
    "Mines": [0, 0, 1, 0, 0, 0, 0],
    "Negative": [0, 0, 0, 1, 0, 0, 0],
    "ProcessingPlants": [0, 0, 0, 0, 1, 0, 0],
    "RefineriesAndTerminals": [0, 0, 0, 0, 0, 1, 0],
    "WWTreatment": [0, 0, 0, 0, 0, 0, 1]
}
```

### Reading Test & Validation data from personal drive after download (Not executable)

### Test Data: Filtering for images labled according to the 7 METER ML labels

```python
labels = ["CAFOs", "Landfills", "Mines", "Negative", "ProcessingPlants", "RefineriesAndTerminals", "WWTreatment"]
```

```python
#read file
personal_path = '/Personal/test_dataset.geojson'

# Read the GeoJSON file into a GeoDataFrame
test = gpd.read_file(personal_path)
test = pd.DataFrame(test.drop(columns='geometry'))
#only keep relevant columns
test = test[["Image_Folder", "Type"]]

#init df
test_final = pd.DataFrame()

#only use files with 
for label in labels:
    label = test[test['Type'] == label]
    test_final = pd.concat([test_final, label])



test_final['Label'] = test_final['Type'].apply(lambda x: mapping[x])
test_final = test_final.drop("Type", axis=1)

test_final['Image_Folder'] = '/samples' + test_final['Image_Folder'].astype(str)

test_final.to_csv('data/FINAL_METER_ML_train.csv', index=False)

test_final
```

### Validation data

```python
#read file
personal_path = '/Users/mauricebaier/Downloads/val_dataset.geojson'

# Read the GeoJSON file into a GeoDataFrame
val = gpd.read_file(personal_path)
val = pd.DataFrame(val.drop(columns='geometry'))
#only keep relevant columns
val = val[["Image_Folder", "Type"]]

#init df
val_final = pd.DataFrame()

#only use files with 
for label in labels:
    label = val[val['Type'] == label]
    val_final = pd.concat([val_final, label])



val_final['Label'] = val_final['Type'].apply(lambda x: mapping[x])
val_final = val_final.drop("Type", axis=1)

val_final['Image_Folder'] = '/samples' + val_final['Image_Folder'].astype(str)

val_final.to_csv('data/FINAL_METER_ML_val.csv', index=False)

val_final
```

### Downsampling even further due to later decision on magnitude of dataset: max. 2000 instances per class

```python
df = pd.read_csv('METER_ML_train_downsampled.csv')
labels = ["CAFOs","Landfills","Mines","Negative","ProcessingPlants","RefineriesAndTerminals","WWTreatment"]

df_2000 = pd.DataFrame()

for label in labels:
    df_label = df[df["Type"] == label]
    df_label = df_label.head(2000)
    df_2000 = pd.concat([df_2000, df_label], ignore_index=True)

df_2000
```

### Reducing previously created augmentation of images due to decision for fewer training data

```python
pp_aug = pd.read_csv('data/save/pp.csv')
pp_aug = pp_aug.head(170)


mines_aug = pd.read_csv('data/save/mines.csv')
mines_aug = mines_aug.head(294)

df_2000 = pd.concat([df_2000, pp_aug, mines_aug], ignore_index=True)
df_2000 = df_2000.drop('Unnamed: 0', axis=1)  # axis=1 specifies that we are dropping a column

df_2000
```
